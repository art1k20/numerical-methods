```python
A = [3.81 0.25 1.28 1.75; % Матриця системи A
 2.25 1.32 5.58 0.49;
 5.31 7.28 0.98 1.04;
 10.39 2.45 3.35 2.28]
 
 b = [4.21 8.97 2.38 12.98]'
 
```

```python
B = A' * A
 
 d = A' * b
 
  B([5 9 10  13 14 15]) = 0
  
  L = B
 B = A' * A
 B([1 2 3 4 6 7 8 11 12 16]) = 0
 U = B
  B = A' * A
 T = -(L \ U)
 
 C = L \ d;
```

```python

 x_kp1 = zeros(4, 1); x_k = ones(4, 1);
epsilon = 0.00001; % Точність
i = 0; % Лічильник ітерацій
while max(abs(x_kp1 - x_k)) > epsilon
 x_k = x_kp1;
 x_kp1 = T * x_k + C;
 i = i + 1;
end
```

```python
format long
fprintf('Approximate solution with precision epsilon = %d found in %u iterations : ', epsilon, i ) 
x = x_k

fprintf('Absolute error vector = (Ax - b) : '); err_vec = A*x -b
fprintf('Absolute error magnitude = |Ax - b| : '); err_abs = norm(A*x - b)
```
